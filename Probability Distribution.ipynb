**bernouli distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Define probability of success
p = 0.6

# Generate 1000 Bernoulli samples
data = bernoulli.rvs(p, size=1000)

# Plot the distribution
plt.bar([0, 1], [sum(data == 0), sum(data == 1)], color=['pink', 'grey'], tick_label=["Failure (0)", "Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

**binomial distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # Number of dice rolls
p = 0.3  # Probability of rolling a 6

# Possible outcomes: 0 to n (number of times we roll a 6)
x = np.arange(0, n + 1)

# Compute Binomial PMF values
pmf_values = binom.pmf(x, n, p)

# Plot the PMF
plt.bar(x, pmf_values, color='pink', alpha=0.7)
plt.xlabel('Number of times rolling a 6')
plt.ylabel('Probability Mass Function (PMF)')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.show()
```

**poission distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter (λ): Average number of events per interval
λ = 3  

# Possible number of occurrences (0 to 10)
x = np.arange(0, 10)

# Compute Poisson PMF values
pmf_values = poisson.pmf(x, λ)

# Plot the PMF
plt.bar(x, pmf_values, color='darkcyan',alpha=0.7)
plt.xlabel('Number of calls per Minute')
plt.ylabel('Probability ')
plt.title(f'Poisson Distribution (λ={λ})')
plt.show()
```

**normal disrtibution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters
mu = 70    # Mean
sigma = 10 # Standard deviation

# Generate Normal distribution data
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)  # Range of scores
pdf = norm.pdf(x, mu, sigma)  # Compute PDF

# Plot the Normal Distribution
plt.plot(x, pdf, color='purple', label=f"N({mu}, {sigma}²)")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (μ={mu}, σ={sigma})")
plt.legend()
plt.grid()
plt.show()
```

**standard normal distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate standard normal distribution data
z = np.linspace(-4, 4, 1000)  # Z-values from -4 to 4
pdf = norm.pdf(z, 0, 1)  # Standard Normal Distribution (μ=0, σ=1)

# Plot the Standard Normal Distribution
plt.plot(z, pdf, color='blue', label="Standard Normal Distribution (μ=0, σ=1)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.grid()
plt.show()
```

**central limit theorem**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate a skewed income dataset (exponential distribution)
income_data = np.random.exponential(scale=50, size=1000)

# Plot the original skewed distribution
plt.figure(figsize=(8, 4))
sns.histplot(income_data, bins=50, kde=True, color="r")
plt.title("Original Income Distribution (Skewed)")
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()

# Perform sampling and calculate means
sample_means = []
n_samples = 1000  # Number of samples
sample_size = 30  # Sample size

for _ in range(n_samples):
    sample = np.random.choice(income_data, sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=50, kde=True, color="b")
plt.title("Sampling Distribution of the Mean")
plt.xlabel("Sample Mean Income ($1000)")
plt.ylabel("Frequency")
plt.show()
```
